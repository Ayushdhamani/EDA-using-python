```python
#Loading the required Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('fifa.csv')
fifa.head()
```

```python
for col in fifa.columns:
    print(col)
```

```python
fifa['nationality'].value_counts()
```

```python
##Getting the top 5 Countries
fifa['nationality'].value_counts()[0:10]
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()),list(fifa['nationality'].value_counts()[0:10]),color='purple')
plt.xlabel('Nationality')
plt.ylabel('No. of Players')
plt.grid(True)
plt.show()
```

```python
#Players salary in euro with name
player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
#Top 10 world's highest wage paid players
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(player_salary['short_name'])[0:10],list(player_salary['wage_eur'][0:10]),color='g')
plt.xlabel('Players Name')
plt.ylabel('Wages')
plt.grid(True)
plt.show()
```
